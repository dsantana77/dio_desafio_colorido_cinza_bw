```python
# Bibliotecas
from google.colab import files
import cv2
import os
import matplotlib.pyplot as plt
```

```python
# Etapa 1: Upload da imagem
uploaded = files.upload()

# Etapa 2: Verifica se algo foi enviado
if not uploaded:
    print("Nenhum arquivo foi enviado.")
    exit()
```

```python
# Etapa 3: Pega o nome do arquivo enviado
nome_arquivo = list(uploaded.keys())[0]

# Etapa 4: Lê a imagem com OpenCV
imagem_colorida = cv2.imread(nome_arquivo)

if imagem_colorida is None:
    print("Erro ao carregar a imagem.")
    exit()
```

```python
# Converte de BGR para RGB para exibir corretamente com matplotlib
imagem_rgb = cv2.cvtColor(imagem_colorida, cv2.COLOR_BGR2RGB)

# Etapa 5: Converte para escala de cinza
imagem_cinza = cv2.cvtColor(imagem_colorida, cv2.COLOR_BGR2GRAY)

# Etapa 6: Binarização (limiar 127)
_, imagem_binarizada = cv2.threshold(imagem_cinza, 127, 255, cv2.THRESH_BINARY)
```

```python
# Etapa 7: Exibe as imagens
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.title("Imagem Colorida")
plt.imshow(imagem_rgb)
plt.axis("off")

plt.subplot(1, 3, 2)
plt.title("Imagem em Tons de Cinza")
plt.imshow(imagem_cinza, cmap='gray')
plt.axis("off")

plt.subplot(1, 3, 3)
plt.title("Imagem Binarizada")
plt.imshow(imagem_binarizada, cmap='gray')
plt.axis("off")

plt.tight_layout()
plt.show()
```

```python
# Etapa 8: Salva as imagens no mesmo diretório
nome_base = os.path.splitext(nome_arquivo)[0]
cv2.imwrite(f"{nome_base}_cinza.png", imagem_cinza)
cv2.imwrite(f"{nome_base}_bin.png", imagem_binarizada)

print("✅ Imagens salvas com sucesso!")
```
